```python
def interroger_API_MovieDB(id_acteur):
    url = f"""https://api.themoviedb.org/3/discover/movie?sort_by=popularity.desc&language=fr-FR
              &with_cast={id_acteur}&api_key=eb2a8d862b4596766c403f227c54b86f"""
    reponse = requests.get(url)
    reponse = reponse.json()
    nb_pages = reponse['total_pages']
    tab = []
    for num_page in range(1, 4):
        url = f"""https://api.themoviedb.org/3/discover/movie?sort_by=popularity.desc&language=fr-FR
                  &with_cast={id_acteur}&page={num_page}&api_key=eb2a8d862b4596766c403f227c54b86f"""
        reponse = requests.get(url)
        reponse = reponse.json()
        for elem in reponse['results']:
            tab.append(elem)
    return tab
```
