# API Vaccination COVID

```python
import requests
```

```python
def interroger_API_vaccination(dep):
    url = f"https://public.opendatasoft.com/api/records/1.0/search/?dataset=covid19-france-lieu-vaccination&q=&rows=100&refine.dep_code={dep}"
    reponse = requests.get(url)
    reponse = reponse.json()
    tab = [{'nom' : elem['fields']['name'],
            'pos' : elem['fields']['geo_point_2d']
    } for elem in reponse['records']]
    return tab
```

```python
donnees = interroger_API_vaccination(94)
```

```python
import folium
```

```python
carte = folium.Map(location = (48.79, 2.42), tiles = 'OpenStreetMap', zoom_start = 12)
for k in range(len(donnees)):
    folium.Marker(donnees[k]['pos'],  tooltip = donnees[k]['nom']).add_to(carte)
carte
```
